```python
print('Whether a number is Positive or Negative?')
number = int(input('Enter your number: '))
if number> 0 :
    print(f'The number is positive')
elif number == 0 :
    print('The number is Ziro')
else :
    print('The number is Negative')
```

```python
print('Whether a number is Even or Odd?')
number = int(input('Enter your number: '))
if number % 2 == 0 :
    print(f'The number is Even')
else: 
    print('The number is Odd')
```

```python
print('Whether a year is Leap year or not?')
year = int(input('Enter your Year: '))
if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0 :
    print(f'Yes, the {year} is Leap Year.')
else :
    print(f'No, the {year} is not Leap Year.')
```

```python
print('Whether a number is Prime or not?')
number = int(input('Enter your number: '))
Prime = 'Prime'
for i in range(2, number // 2) :
    if number % i ==  0 :
        Prime = 'Not Prime'
        break
print(f'The number {number} is {Prime}')
```

```python
print('Print all Prime between 1-10000?')
for j in range(1 , 10000): 
    Prime = True
    for i in range(2, j // 2) :
        if j % i ==  0 :
            Prime = False
            break
    if Prime: print(f'The number {j} is Prime')
```
